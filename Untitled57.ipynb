```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

import pickle
```

```python
df = pd.read_csv("/content/weatherHistory.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

### Drop unnecessary Colomns

```python
df.drop(['Loud Cover', 'Daily Summary'], axis=1, inplace=True)
```

### Handle Missing Values

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

### Date-Time Feature

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], errors='coerce')
df = df.dropna(subset=['Formatted Date'])
df['Month'] = df['Formatted Date'].dt.month
df['Hour'] = df['Formatted Date'].dt.hour
```

### Target Lable

```python
def severity_label(summary):
    summary = summary.lower()
    if any(word in summary for word in ['thunderstorm', 'heavy rain', 'hail', 'sleet']):
        return 'Severe'
    elif any(word in summary for word in ['rain', 'snow', 'foggy', 'windy']):
        return 'Mild'
    else:
        return 'Normal'

df['Severity'] = df['Summary'].apply(severity_label)
```

```python
le = LabelEncoder()
df['Precip Type'] = le.fit_transform(df['Precip Type'])
```

### Features and Labels

```python
features = ['Temperature (C)', 'Apparent Temperature (C)', 'Humidity',
            'Wind Speed (km/h)', 'Wind Bearing (degrees)', 'Visibility (km)',
            'Pressure (millibars)', 'Precip Type', 'Month', 'Hour']

X = df[features]
y = df['Severity']
```

```python
y_encoded = le.fit_transform(y)  # Encode 'Normal', 'Mild', 'Severe' as 0,1,2
```

### Split Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

### Scaling

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Use random forest classifier

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

### Evaluation

```python
y_pred = model.predict(X_test)
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=le.classes_))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

y_pred = model.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=le.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```

```python
with open("extreme_weather_model.pkl", "wb") as f:
    pickle.dump(model, f)

with open("label_encoder.pkl", "wb") as f:
    pickle.dump(le, f)

with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)
```
